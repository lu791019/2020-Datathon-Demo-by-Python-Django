```python
df_rfm['freq_lt_z'] = (df_rfm['frequency_lt'] - df_rfm['frequency_lt'].mean())/df_rfm['frequency_lt'].std()
df_rfm['freq_3_z'] = (df_rfm['frequency_3days'] - df_rfm['frequency_3days'].mean())/df_rfm['frequency_3days'].std()
df_rfm['freq_7_z'] = (df_rfm['frequency_7days'] - df_rfm['frequency_7days'].mean())/df_rfm['frequency_7days'].std()
df_rfm['freq_14_z'] = (df_rfm['frequency_14days'] - df_rfm['frequency_14days'].mean())/df_rfm['frequency_14days'].std()
df_rfm['freq_30_z'] = (df_rfm['frequency_30days'] - df_rfm['frequency_30days'].mean())/df_rfm['frequency_30days'].std()

df_rfm['freq_lt_p'] = scipy.stats.norm.sf(abs(df_rfm.freq_lt_z))
df_rfm['freq_3_p'] = scipy.stats.norm.sf(abs(df_rfm.freq_3_z))
df_rfm['freq_7_p'] = scipy.stats.norm.sf(abs(df_rfm.freq_7_z))
df_rfm['freq_14_p'] = scipy.stats.norm.sf(abs(df_rfm.freq_14_z))
df_rfm['freq_30_p'] = scipy.stats.norm.sf(abs(df_rfm.freq_30_z))

df_rfm['weather'] = 1
```
